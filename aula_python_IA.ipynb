```python
media = float(input("Digite a sua média: "))

if media >= 7.0:
    print("Aprovado")
elif media >= 5.0:
    print("Recuperação")
else:
    print("Reprovado")
```

```python
media = float(input("Digite a sua média: "))

if media < 5.0:
    print("Reprovado")
elif media >= 5.0 and media < 7.0:
    print("Recuperação")
else:
    print("Aprovado")
```
